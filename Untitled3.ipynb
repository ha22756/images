```python
تمام شوف هدا الكود
وعدلة ليا مايناسب المطلوب وخليك مركز نبيهن يطلعن وينعرضن الفلاتر وخليهن كويسات

import numpy as np
import matplotlib.pyplot as plt
import requests
from io import BytesIO
from PIL import Image
from scipy import ndimage as ndi
from skimage import transform
from skimage.feature import peak_local_max
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D

# ==========================================================
# 1. روابط الملفات (استخدم الروابط الفعلية لديك)
# ==========================================================
IMG_URL = "https://raw.githubusercontent.com/ha22756/images/main/JPCLN001%20(1).png"
FULL_MASK_URL = "https://raw.githubusercontent.com/ha22756/images/main/JPCLN001.png"  # القناع الواحد المدمج

def load_image(url):
    resp = requests.get(url)
    resp.raise_for_status()
    return np.array(Image.open(BytesIO(resp.content)).convert('L'))

print(">>> تحميل الصورة والقناع الكامل...")
img = load_image(IMG_URL)
full_mask = load_image(FULL_MASK_URL)

# ==========================================================
# 2. تجهيز الصورة (Resampling & LCE)
# ==========================================================
target_size = (512, 512)
img_res = transform.resize(img, target_size, preserve_range=True).astype(np.float32)
mask_res = transform.resize(full_mask, target_size, order=0) > 0.5   # قناع موحد (الرئتين معاً)

# Local Contrast Enhancement
gauss_smooth = ndi.gaussian_filter(img_res, sigma=16)
gauss_sq = ndi.gaussian_filter(img_res**2, sigma=16)
lce = (img_res - gauss_smooth) / (np.sqrt(gauss_sq - gauss_smooth**2) + 1e-6)

# ==========================================================
# 3. بناء فلاتر WMCI الأسطوانية
# ==========================================================
def create_stepped_kernel(outer_diameter):
    radius = outer_diameter / 2.0
    size = 31
    kernel = np.zeros((size, size))
    center = size // 2
    y, x = np.ogrid[-center:center+1, -center:center+1]
    r_dist = np.sqrt(x**2 + y**2)
    max_w = 0.008
    kernel[r_dist <= radius * 0.5] = max_w
    kernel[(r_dist > radius * 0.5) & (r_dist <= radius * 0.8)] = max_w * 0.6
    kernel[(r_dist > radius * 0.8) & (r_dist <= radius)] = max_w * 0.3
    return kernel

w1 = create_stepped_kernel(14)   # 9.8 mm
w2 = create_stepped_kernel(20)   # 14 mm
w3 = create_stepped_kernel(26)   # 18.2 mm

# Convolution
print(">>> إجراء الـ Convolution...")
conv1 = ndi.convolve(lce, w1)
conv2 = ndi.convolve(lce, w2)
conv3 = ndi.convolve(lce, w3)
combined_response = (conv1 + conv2 + conv3) / 3.0

# تطبيق القناع الكامل (الرئتين معاً)
masked_response = np.where(mask_res, combined_response, 0)

# ==========================================================
# 4. المهمة الأولى: عرض أقنعة منفصلة (تمثيلية) + القناع المدمج الحقيقي
# ==========================================================
# توليد قناعين افتراضيين (يسار ويمين) من القناع الكامل لعرضهما فقط
h, w = mask_res.shape
mid = w // 2
left_mask = mask_res.copy()
left_mask[:, mid:] = False
right_mask = mask_res.copy()
right_mask[:, :mid] = False

fig, axes = plt.subplots(1, 4, figsize=(16, 5))
axes[0].imshow(img_res, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(left_mask, cmap='gray')
axes[1].set_title("Left Mask (simulated)")
axes[1].axis('off')

axes[2].imshow(right_mask, cmap='gray')
axes[2].set_title("Right Mask (simulated)")
axes[2].axis('off')

axes[3].imshow(mask_res, cmap='gray')
axes[3].set_title("Combined Mask (actual)")
axes[3].axis('off')

plt.suptitle("Note: Left/Right masks are generated from the full mask for display only.", fontsize=10)
plt.tight_layout()
plt.show()

# حفظ القناع المدمج (نفس القناع الكامل)
Image.fromarray((mask_res * 255).astype(np.uint8)).save("combined_lung_mask.png")
print("✔ تم حفظ القناع المدمج باسم combined_lung_mask.png")

# ==========================================================
# 5. المهمة الثانية: عرض مراحل المعالجة + الاكتشافات النهائية
# ==========================================================
fig, axes = plt.subplots(1, 4, figsize=(18, 5))
axes[0].imshow(img_res, cmap='gray')
axes[0].set_title("1. Resampled Image (512x512)")
axes[0].axis('off')

axes[1].imshow(lce, cmap='gray')
axes[1].set_title("2. Local Contrast Enhancement")
axes[1].axis('off')

axes[2].imshow(masked_response, cmap='hot')
axes[2].set_title("3. Response after Lung Mask")
axes[2].axis('off')

# الاكتشافات بعتبة 0.45 (يمكن تغييرها)
threshold_val = 0.45
peaks = peak_local_max(masked_response, min_distance=12,
                       threshold_abs=masked_response.max() * threshold_val)

axes[3].imshow(img_res, cmap='gray')
axes[3].contour(mask_res, colors='cyan', linewidths=1.2)   # قناع الرئة الموحد
axes[3].scatter(peaks[:, 1], peaks[:, 0], s=150, edgecolors='red',
                facecolors='none', linewidths=2)
axes[3].scatter(peaks[:, 1], peaks[:, 0], marker='x', color='red', s=40)
axes[3].set_title(f"4. Final Detections (Thr={threshold_val}, Count={len(peaks)})")
axes[3].axis('off')

plt.tight_layout()
plt.show()

# ==========================================================
# 6. رسم الفلاتر ثلاثية الأبعاد
# ==========================================================
fig = plt.figure(figsize=(18, 6))
X, Y = np.meshgrid(np.arange(31)-15, np.arange(31)-15)
kernels = [w1, w2, w3]
titles = ['(a) 14 px (9.8 mm)', '(b) 20 px (14 mm)', '(c) 26 px (18.2 mm)']

for i in range(3):
    ax = fig.add_subplot(1, 3, i+1, projection='3d')
    ax.plot_surface(X, Y, kernels[i], cmap='plasma', edgecolor='none')
    ax.set_title(f"Weighting Function {titles[i]}")
    ax.set_zlim(0, 0.01)
    ax.set_xlabel('k')
    ax.set_ylabel('l')
    ax.set_zlabel('w(k,l)')
plt.show()

# ==========================================================
# 7. المهمة الثالثة: تحليل تأثير العتبة
# ==========================================================
thresholds = np.arange(0.30, 0.65, 0.05)  # 0.30, 0.35, ..., 0.60
counts = []
for thr in thresholds:
    p = peak_local_max(masked_response, min_distance=12,
                       threshold_abs=masked_response.max() * thr)
    counts.append(len(p))

# جدول
df = pd.DataFrame({'Threshold': thresholds, 'Detected Candidates': counts})
print("\n--- نتائج تحليل العتبة (المهمة الثالثة) ---")
print(df.to_string(index=False))

# رسم بياني
plt.figure(figsize=(8, 5))
plt.plot(thresholds, counts, marker='o', linestyle='-', color='b')
plt.xlabel('Threshold Value')
plt.ylabel('Number of Candidates')
plt.title('Effect of Threshold on WMCI Detections')
plt.grid(True, alpha=0.3)
plt.show()

# تعليق قصير
print("""
التعليق:
- كلما زادت قيمة العتبة يقل عدد المرشحين لأننا نطلب استجابة أقوى لاعتبارها اكتشافاً محتملاً.
- العلاقة عكسية غير خطية؛ عند العتبات المنخفضة جداً (0.30) يظهر عدد كبير من النقاط (معظمها إيجابيات كاذبة)،
  بينما عند العتبات المرتفعة (0.60) يتبقى فقط أبرز التموضعات (غالباً عقيدات حقيقية).
- أنسب نطاق عملي يقع بين 0.40 و 0.50 حيث يتوازن الحساسية مع تقليل الضوضاء.
""")
```
